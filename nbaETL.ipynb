```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_path = "20182019.csv"
nba_df = pd.read_csv(csv_path)
```

```python
nba_df.columns
```

```python
nba_df['Player'] = nba_df['Player'].str.split('\\').str[0]

nba_df = nba_df.drop(columns=['Rk'])
```

```python
nba_df = nba_df.rename(columns={'FG%': 'FG Percentage', 'FG%': 'FG Percentage', '3P%': '3P Percentage', '2P%': '2P Percentage', 'eFG%': 'eFG Percentage', 'FT%': 'FT Percentage'})

nba_df.columns
```

```python
nba_df = nba_df.drop_duplicates(subset='Player', keep='first')

nba_df.head(30)
```

```python
connection_string = "postgres:parker@localhost:5432/nba_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
nba_df.to_sql(name='2018-2019 Season Stats', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from nba_db', con=engine).head()
```
